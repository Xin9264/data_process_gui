```python
import PIL
from PIL import Image
from GPT_response import gpt4_chat_2images
from cut_keyframe import parse_sequence_file
import matplotlib.pyplot as plt
```

阅读sequence.txt文件，获得对应的动作

```python
actions = parse_sequence_file(sequence_file="./example/sequence.txt")
actions[289]
```

第二张图片的名称数字对应actions中的动作

比如第二张图片是289结尾的，对应的动作就是actions[289]

```python
image1 = Image.open("example/frame_0288.jpg")
image2 = Image.open("example/frame_0289.jpg")
sequence_index = 289
response = gpt4_chat_2images(image1, image2, action=actions[sequence_index][1:])
```

```python
plt.figure(figsize=(200, 400))
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.axis('off')
```
